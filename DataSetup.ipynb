```python
def percent_to_decimal(string: str) -> float:
    num_str = ""
    for char in string:
        if char != "%":
            num_str += char
        else:
            break
    num = float(num_str)
    return num/100
```

```python
def list_percent_to_decimal(string_list: list) -> list:
    new_list = []
    for elem in string_list:
        new_list.append(percent_to_decimal(elem))
    return new_list
```

```python
lefty_df["BB%"] = list_percent_to_decimal(list(lefty_df["BB%"]))
lefty_df["K%"] = list_percent_to_decimal(list(lefty_df["K%"]))
righty_df["BB%"] = list_percent_to_decimal(list(righty_df["BB%"]))
righty_df["K%"] = list_percent_to_decimal(list(righty_df["K%"]))
```

```python
def swap_name(string_list: list) -> list:
    """Given a list of names separated by commas, reformat them from last, first to first last"""
    newList = list([])
    for string in string_list:
        first = ""
        last = ""
        i = 0
        while string[i:i + 1] != ",":
            last += string[i:i + 1]
            i += 1
        i += 1
        while i < len(string):
            first += string[i:i + 1]
            i += 1
        newList.append(first.strip()+" "+last.strip())
    return newList
```
